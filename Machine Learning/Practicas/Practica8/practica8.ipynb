```python
#importamos las librerias necesarias
import tensorflow as tf#libreria de google para el manejo de redes neuronales
import numpy as np  #libreria para el manejo de arreglos
```

```python
#Damos los datos de entrada y salida para el modelo
celsius = np.array([ 0, -6.66667 , -9.44444 , 10 , 18.3333 , 22.2222 , -15, 29.4444, 37.7778, -17.7778 ], dtype=float)#creamos un array con los valores de celsius
fahrenheit = np.array([32, 20, 15, 50, 65, 72, 5, 85, 100, 0], dtype=float)#creamos un array con los valores de fahrenheit
```

```python
#creamos el modelo
capa = tf.keras.layers.Dense(units=1, input_shape=[1])#creamos una capa con una sola neurona
modelo = tf.keras.Sequential([capa])#creamos el modelo con la capa creada
```

```python
#compilamos el modelo
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),#usamos el optimizador Adam con un ratio de aprendizaje de 0.1
    loss='mean_squared_error'#usamos la funcion de perdida de error cuadratico medio
)
```

```python
#entrenamos el modelo
print("Comenzando entrenamiento...")#imprimimos un mensaje para saber que el entrenamiento ha comenzado
historial = modelo.fit(celsius, fahrenheit, epochs=1000, verbose=False)#entrenamos el modelo con los datos de entrada y salida
print("Modelo entrenado!")#imprimimos un mensaje para saber que el entrenamiento ha terminado
```

```python
import matplotlib.pyplot as plt#libreria para graficar
plt.xlabel("# Epoca")#etiqueta del eje x
plt.ylabel("Magnitud de pérdida")#etiqueta del eje y
plt.plot(historial.history["loss"])#graficamos la perdida
```

```python
print("Hagamos una predicción!")#imprimimos un mensaje para saber que vamos a hacer una prediccion
resultado = modelo.predict([100.0])#hacemos una prediccion con el modelo
print("El resultado es " + str(resultado) + " fahrenheit!")#imprimimos el resultado de la prediccion
```
